### Importing Libraries

```python
import pandas as pd
import os
```

###  Defining the folder path where the CSV files are located



```python
folder_path = '/home/shirleen/Documents/QGIS Zoning Project/Animal Observations'
```

```python

dfs = []

# Iterating through the CSV files in the folder
for csv_file in os.listdir(folder_path):
    if csv_file.endswith('.csv'):
        file_path = os.path.join(folder_path, csv_file)

        # Loading the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Identifying the relevant columns for region and observation counts
        region_column = 'Name'
        observation_count_column = 'Type of observation_count'

        df.rename(columns={region_column: 'Region', observation_count_column: 'Observation_Count'}, inplace=True)

        # Appending the DataFrame to the list
        dfs.append(df)

# Concatenating all DataFrames in the list
combined_df = pd.concat(dfs, ignore_index=True)

# Grouping the combined DataFrame by both 'Region' and 'Animal' and calculating the sum of observation counts
aggregated_df = combined_df.groupby(['Region', 'Animal'])['Observation_Count'].sum().reset_index()

aggregated_df
    
```

```python
# Grouping the aggregated_df by 'Region' and calculate the sum of observation counts for each region
total_counts_by_region = aggregated_df.groupby('Region')['Observation_Count'].sum().reset_index()

# Renaming the columns for clarity (optional)
total_counts_by_region.rename(columns={'Observation_Count': 'Total_Observation_Count'}, inplace=True)

total_counts_by_region
```
